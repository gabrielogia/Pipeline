```python
import pandas as pd
```

```python
df_nm = pd.read_csv('reduced_molecules/7294/spec-nm.dat', sep=' ', names=['a', 'b'])
df_ev = pd.read_csv('reduced_molecules/7294/spec-ev.dat', sep=' ', names=['a', 'b'])
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(20,5))
plt.plot(df_nm['a'], df_nm['b'])
plt.xlabel('Wavelength/nm')
plt.ylabel('Absorbance [abs. units]')
plt.xlim([df_nm['a'].min(), 600])
plt.xticks(np.linspace(df_nm['a'].min(), 600, 10))
plt.title('Absorbance - Methyl acrylate')
plt.show()
```

```python
plt.figure(figsize=(20,5))
plt.plot(df_ev['a'], df_ev['b'])
plt.xlabel('e.v.')
plt.ylabel('Absorbance')
plt.title('Absorbance - Methyl acrylate')
plt.show()
```
